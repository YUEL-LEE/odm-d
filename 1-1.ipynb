```python
# 모듈 불러오기
import  pandas as pd
import matplotlib.pyplot as plt

# 긴 데이터 링크를 data_url로 치환
data_url = "https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx"
```

```python
# 첫 번째 시트, df1 불러오기
df1 = pd.read_excel(data_url, sheet_name = 0)
df1
```

```python
# 두 번째 시트, df2 불러오기
df2 = pd.read_excel(data_url, sheet_name = 1)
df2
```

```python
# row와 column을 tranpose한 데이터를 각각의 df로 설정
df1 = df1.T
df2 = df2.T
```

```python
# 각 시트의 df의 첫 가로 행을 header로 설정
header1 = df1.iloc[0]
df1 = df1[1:]
df1.columns = header1
```
